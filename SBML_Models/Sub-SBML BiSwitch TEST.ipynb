```python
# Imports
#from genelet import*
from subsbml import System, createNewSubsystem, createSubsystem, combineSystems
import numpy as np
import pylab as plt


from bokeh.layouts import row
from bokeh.io import export_png

import warnings
import bokeh.io
import bokeh.plotting
```

# Bistable Switch Recreate

```python
ss1 = createSubsystem ("Bistable_Switch/BiSwitch_CRN.xml")


Core1_id = ss1.getSpeciesByName('complex_Core1_ON').getId()
Core1R_id = ss1.getSpeciesByName('Core1_OFF').getId()
Core2_id = ss1.getSpeciesByName('complex_Core2_ON').getId()
Core2R_id = ss1.getSpeciesByName('Core2_OFF').getId()
dA2_rR2 = ss1.getSpeciesByName("complex_Core1_AI").getId()


# Set species amount
ss1.setSpeciesAmount('Core1_OFF', 6e3)
ss1.setSpeciesAmount('Core2_OFF', 5e3)
ss1.setSpeciesAmount('dA2', 6e3)
ss1.setSpeciesAmount('dA1', 6e3)
ss1.setSpeciesAmount('protein_RNAseH', 5)
ss1.setSpeciesAmount('protein_RNAP', 150)
```

```python
# Simulate with Bioscrape
timepoints = np.linspace(0, 28800, 1000)
results,_ = ss1.simulateWithBioscrape(timepoints)
timepoints = timepoints/3600

# For label convenience
x = 'Time (seconds)'
y = 'Concentration (uM)'

# Core 1 & Core 2
bokeh.io.output_notebook()
a = bokeh.plotting.figure(plot_width=400, plot_height=300, x_axis_label = x, y_axis_label = y)
a.circle(timepoints, results[Core1_id], legend_label = "ON Core 1" , color = "blue")
a.circle(timepoints, results[Core2_id], legend_label = "ON Core 2", color = "red")
a.legend.click_policy="hide"
a.legend.location="bottom_right"


bokeh.io.show(row(a))
warnings.filterwarnings("ignore")
```
