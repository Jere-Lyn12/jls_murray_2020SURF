```python
# Imports
from biocrnpyler import *
from genelet import *
from subsbml import System, createSubsystem, combineSystems, createNewSubsystem, createBasicSubsystem, SimpleModel, SimpleReaction
import numpy as np
import pylab as plt

from bokeh.layouts import row
from bokeh.io import export_png

import warnings
import libsbml
import bokeh.io
import bokeh.plotting
import bioscrape
```

```python
ss1 = createSubsystem('Prod2_CRN.xml')
ss2 = createSubsystem('DFHBI.xml')

Produce_subsystem = ss1 + ss2
_ = Produce_subsystem.writeSBML('Full_Produce.xml')
print('yay')
```

```python
#### Identifiers ####
X_id1 = Produce_subsystem.getSpeciesByName('complex_Prod2_ON').getId()
X_id2 = Produce_subsystem.getSpeciesByName('Prod2_OFF').getId()
X_id3 = Produce_subsystem.getSpeciesByName('DFHBI').getId()
X_id4 = Produce_subsystem.getSpeciesByName('BrocApt').getId()
X_id5 = Produce_subsystem.getSpeciesByName("complex_BrocApt_DFHBI").getId()


#### Set Species ####
Produce_subsystem.setSpeciesAmount('Prod2_OFF', 5e3)
Produce_subsystem.setSpeciesAmount('dna_dAP', 5e3)
Produce_subsystem.setSpeciesAmount('protein_RNAseH', 0)
Produce_subsystem.setSpeciesAmount('protein_RNAP', 200)
Produce_subsystem.setSpeciesAmount('DFHBI', 60000)
print(X_id3)
```

```python
# Simulate with Bioscrape
timepoints = np.linspace(0, 28800, 1000)
results,_ = Produce_subsystem.simulateWithBioscrape(timepoints)
timepoints = timepoints/3600

# For label convenience
x = 'Time (hours)'
y = 'Concentration (uM)'

# Core 1 & Core 2
bokeh.io.output_notebook()
a = bokeh.plotting.figure(plot_width=400, plot_height=300, x_axis_label = x, y_axis_label = y)
a.circle(timepoints, results[X_id1], legend_label = "ON Prod 2" , color = "blue")
a.circle(timepoints, results[X_id2], legend_label = "OFF Prod 2", color = "red")
a.legend.click_policy="hide"
a.legend.location="bottom_right"

b = bokeh.plotting.figure(plot_width=400, plot_height=300, x_axis_label = x, y_axis_label = y)
b.circle(timepoints, results[X_id4], legend_label = "Brocco" , color = "blue")
b.legend.click_policy="hide"
b.legend.location="bottom_right"


bokeh.io.show(row(a, b))
warnings.filterwarnings("ignore")
```
