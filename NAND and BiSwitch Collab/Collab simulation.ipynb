```python
# Imports
from subsbml import System, createSubsystem, combineSystems, createNewSubsystem, createBasicSubsystem, SimpleModel, SimpleReaction
import numpy as np
import pylab as plt

from bokeh.layouts import row
from bokeh.io import export_png

import warnings
import libsbml
import bokeh.io
import bokeh.plotting
```

## Communication of NAND Gate and Bistable Switch ##
Cell_Bi = Bistable Switch

Cell_N = NAND Gate

External environment = actual external

```python
### Bistable Switch ##
ss1 = createSubsystem ("BiCombine.xml")
ss2 = createSubsystem ("external_environment.xml")

# Membrane
mb1 = createSubsystem("BiSwitch_membrane_detailed.xml", membrane = True)

#### Cell Bi ####
cell_Bi =  System('cell_Bi',ListOfInternalSubsystems = [ss1],
                 ListOfExternalSubsystems = [ss2],
                 ListOfMembraneSubsystems = [mb1])

cell_Bi_model = cell_Bi.getModel()

cell_Bi_model.renameCompartments('cell_Bi_external', 'actual_external')

cell_Bi_model.writeSBML("cell_Bi_model.xml")
```

```python
### NAND Gate ####
ss3 = createSubsystem ("NAND_CRN.xml")

# Membrane
mb2 = createSubsystem("NAND_membrane_detailed.xml", membrane = True)

### Cell_N ####
cell_N =  System('cell_N',ListOfInternalSubsystems = [ss3],
                 ListOfExternalSubsystems = [ss2],
                 ListOfMembraneSubsystems = [mb2])

cell_N_model = cell_N.getModel()

cell_N_model.renameCompartments('cell_N_external', 'actual_external')

cell_N_model.writeSBML("cell_N_model.xml")
```

```python
combined_cells = createNewSubsystem()
combined_cells.combineSubsystems([cell_Bi_model, cell_N_model], verbose = True)
#combined_cells.writeSBML('combined_cells.xml')
#combined_cells.changeLog
```

```python
### Set Species Amount for Combined cells ####
combined_cells.setSpeciesAmount('rna_I1_e', 0.9, compartment = 'actual_external')
combined_cells.setSpeciesAmount('rna_I2_e', 0.9, compartment = 'actual_external')
### Cell_Bi ###
combined_cells.setSpeciesAmount('Core1_OFF', 1.2, compartment = 'cell_Bi_internal') #0.1 (Schaft), 1.2
combined_cells.setSpeciesAmount('Core2_OFF', 0.99, compartment = 'cell_Bi_internal') #0.075 (Schaft), 0.99
combined_cells.setSpeciesAmount('dna_dA2', 1.2, compartment = 'cell_Bi_internal') #0.1, 1.2
combined_cells.setSpeciesAmount('dna_dA1', 1.2, compartment = 'cell_Bi_internal') #0.1, 1.2
combined_cells.setSpeciesAmount('protein_RNAseH', 0.0044, compartment = 'cell_Bi_internal')
combined_cells.setSpeciesAmount('protein_RNAP', 0.030, compartment = 'cell_Bi_internal')

#### Cell_N ###
combined_cells.setSpeciesAmount('Sw1_OFF', 1.2, compartment = 'cell_N_internal') #0.1 (Schaft)
combined_cells.setSpeciesAmount('Sw2_OFF', 1.2, compartment = 'cell_N_internal') #0.075 (Schaft)
combined_cells.setSpeciesAmount('Sw3_OFF', 1.2, compartment = 'cell_N_internal') #0.075 (Schaft)

combined_cells.setSpeciesAmount('dna_A1', 1.2, compartment = 'cell_N_internal') #0.1
combined_cells.setSpeciesAmount('dna_A2', 1.2, compartment = 'cell_N_internal') #0.1
combined_cells.setSpeciesAmount('dna_A31', 1.2, compartment = 'cell_N_internal') #0.1
combined_cells.setSpeciesAmount('dna_A32', 1.2, compartment = 'cell_N_internal') #0.1

combined_cells.setSpeciesAmount('protein_RNAseH', 0.0044, compartment = 'cell_N_internal')
combined_cells.setSpeciesAmount('protein_RNAP', 0.030, compartment = 'cell_N_internal')

combined_cells.writeSBML('combined_cells.xml')
combined_cells.changeLog
```

```python
# Cell_Bi Identifiers #

B_id1 = combined_cells.getSpeciesByName('complex_Core1_ON').getId()
B_id2 = combined_cells.getSpeciesByName('Core1_OFF').getId()
B_id3 = combined_cells.getSpeciesByName('complex_Core2_ON').getId()
B_id4 = combined_cells.getSpeciesByName('Core2_OFF').getId()
B_id5 = combined_cells.getSpeciesByName("rna_rR2").getId()
B_id6 = combined_cells.getSpeciesByName("rna_rR1").getId()


B_id8 = combined_cells.getSpeciesByName("complex_Core2_AI").getId()
B_id9 = combined_cells.getSpeciesByName("complex_Core1_AI").getId()
B_id10 = combined_cells.getSpeciesByName("dna_dA2").getId()
B_id11 = combined_cells.getSpeciesByName("dna_dA1").getId()

B_id12 = combined_cells.getSpeciesByName("rna_I4_B", compartment = "cell_Bi_internal").getId()

W_id1 = combined_cells.getSpeciesByName("complex_rna_I4_B_rna_rR1", compartment = "cell_Bi_internal").getId()
W_id2 = combined_cells.getSpeciesByName("complex_rna_I4_B_rna_rR1_E", compartment = "actual_external").getId()
```

```python
# Cell_N Identifiers #

N_id1 = combined_cells.getSpeciesByName('complex_Sw1_ON').getId()
N_id2 = combined_cells.getSpeciesByName('Sw1_OFF').getId()

N_id3 = combined_cells.getSpeciesByName('complex_Sw2_ON').getId()
N_id4 = combined_cells.getSpeciesByName('Sw2_OFF').getId()

N_id5 = combined_cells.getSpeciesByName('complex_Sw3_ON_1').getId()
N_id6 = combined_cells.getSpeciesByName('complex_Sw3_ON_2').getId()
N_id7 = combined_cells.getSpeciesByName('Sw3_OFF').getId()

N_id8 = combined_cells.getSpeciesByName('rna_I1', compartment = "cell_N_internal").getId()
N_id9 = combined_cells.getSpeciesByName('rna_I2', compartment = "cell_N_internal").getId()
N_id10 = combined_cells.getSpeciesByName('rna_I4', compartment = "cell_N_internal").getId()

N_id11 = combined_cells.getSpeciesByName('rna_I31', compartment = "cell_N_internal").getId()
N_id12 = combined_cells.getSpeciesByName('rna_I32', compartment = "cell_N_internal").getId()


E_id1 = combined_cells.getSpeciesByName('rna_I4_e').getId()

print (E_id1)
```

```python
# Simulate with Bioscrape #
timepoints = np.linspace(0,28800,1000) #86400 (24 hours) or 28800 (8 hours)
R,_ = combined_cells.simulateWithBioscrape(timepoints)
timepoints = timepoints/3600

# For label convenience
x = 'Time (hours)'
y = 'Concentration (uM)'

bokeh.io.output_notebook()
a = bokeh.plotting.figure(title="Bistable Switch",plot_width=400, plot_height=300, x_axis_label = x, y_axis_label=y)
a.circle(timepoints, R[B_id1], legend_label = "Core1 ON" , color = "blue")
a.circle(timepoints, R[B_id3], legend_label = "Core2 ON", color = "red")
a.legend.click_policy="hide"
a.legend.location="bottom_right"

b = bokeh.plotting.figure(title="I4 Species",plot_width=400, plot_height=300, x_axis_label = x, y_axis_label=y)
b.circle(timepoints, R[E_id1], legend_label = "external", color = "orange")
b.circle(timepoints, R[N_id10], legend_label = "NAND", color = "red")
b.circle(timepoints, R[B_id12], legend_label = " BiSwitch", color = "blue")
b.legend.click_policy="hide"
b.legend.location="bottom_right"

c = bokeh.plotting.figure(title="Waste",plot_width=400, plot_height=300, x_axis_label = x, y_axis_label=y)
c.circle(timepoints, R[W_id1], legend_label = "External" , color = "orange")
c.circle(timepoints, R[W_id2], legend_label = "Internal", color = "blue")
c.legend.click_policy="hide"
c.legend.location="bottom_right"

bokeh.io.show(row(a, b, c))
warnings.filterwarnings("ignore")
```

```python
### Simulate NAND Gate ####
d = bokeh.plotting.figure(plot_width=400, plot_height=300, x_axis_label = x, y_axis_label=y)
d.circle(timepoints, R[N_id1], legend_label = "Sw1 ON" , color = "orange")
d.circle(timepoints, R[N_id2], legend_label = "Sw1 OFF", color = "red")
d.legend.click_policy="hide"
d.legend.location="bottom_right"

e = bokeh.plotting.figure(plot_width=400, plot_height=300, x_axis_label = x, y_axis_label=y)
e.circle(timepoints, R[N_id3], legend_label = "Sw1 ON" , color = "orange")
e.circle(timepoints, R[N_id4], legend_label = "Sw1 OFF", color = "red")
e.legend.click_policy="hide"
e.legend.location="bottom_right"

f = bokeh.plotting.figure(plot_width=400, plot_height=300, x_axis_label = x, y_axis_label=y)
f.circle(timepoints, R[N_id5]+R[N_id6], legend_label = "Sw3 ON" , color = "orange")
f.circle(timepoints, R[N_id7], legend_label = "Sw3 OFF", color = "red")
f.legend.click_policy="hide"
f.legend.location="bottom_right"

bokeh.io.show(row(d, e, f))
warnings.filterwarnings("ignore")
```

```python
## Simulate inducer species ##
g = bokeh.plotting.figure(title="Inducers",plot_width=400, plot_height=300, x_axis_label = x, y_axis_label=y)
g.circle(timepoints, R[N_id11], legend_label = "I31", color = "orange")
g.circle(timepoints, R[N_id12], legend_label = "I32", color = "red")
g.circle(timepoints, R[N_id10], legend_label = "I4", color = "green")
g.legend.click_policy="hide"
g.legend.location="bottom_right"

bokeh.io.show(row(g))
warnings.filterwarnings("ignore")
```
